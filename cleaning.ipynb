```python
df = df.drop(['customerID'],axis=1)
```

```python
columns_to_encode =df.drop(['TotalCharges', 'tenure', 'MonthlyCharges'],axis=1).columns
columns_to_encode
```

```python
encoders = {}

for column in columns_to_encode:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
    encoders[column] = encoder
df
```

```python
df.columns[0]
```

```python
encoders['gender'].transform(pd.Series('Male'))
encoders.keys()
```

```python
with open('models/encoders.pkl', 'wb') as f:
    pickle.dump(encoders, f)
```
